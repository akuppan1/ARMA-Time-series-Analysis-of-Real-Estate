```python
#Log transform the data

logged_df = df['SALE PRICE'].apply(lambda x : np.log(x))
logged_df.dropna(inplace=True)
logged_df.reset_index(drop=True)
```

```python
ax1 = plt.subplot(121)
logged_df.plot(figsize=(12,4) ,color="tab:green", title="Log Transformed Values", ax=ax1);
ax2 = plt.subplot(122)
df['SALE PRICE'].plot(color="tab:blue", title="Original Values", ax=ax2);
```

```python
#TEST STATISTIC IS NOT LESS THAN CRITICAL VALUES, WE NEED TO MAKE IT SO!
#differencing the data for stationarity
#Differencing the data simply means to take the difference between the data points and its lagged version.
#I will try a lag of 4

logged_df_diff= logged_df.diff(periods=365)
logged_df_diff.dropna(inplace=True)
logged_df_diff.reset_index(drop=True)
```

```python
dftest = adfuller(logged_df_diff)
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dftest)
print()
print(dfoutput)
```

```python
#ACF and PACF with the vanilla data does not have a unit root and is stationary.
#We have no clear cut correlations
#PACF AND ACF
#Statsmodels ACF
rcParams['figure.figsize'] = 25, 15
plot_acf(logged_df_diff, lags=200, alpha=0.05);

rcParams['figure.figsize'] = 20, 15
plot_pacf(logged_df_diff, lags=200, alpha=0.05);
```

```python
#Subtract the rolling mean from the values, pacf shows me there might be a discernable pattern to express

df1_rolling = logged_df_diff.rolling(window=365).mean()
df1_rolling_mean = logged_df_diff - df1_rolling
df1_rolling_mean.dropna(inplace=True)
```

```python
dftest = adfuller(df1_rolling_mean, maxlag=50)
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dftest)
print()
print(dfoutput)
```
